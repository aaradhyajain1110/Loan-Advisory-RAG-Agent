#  Loan Advisory RAG Pipeline

### A retrieval-augmented generation (RAG) system for answering Indian bank loan questions

**What this notebook does, end to end:**
1. Ingests PDF policy documents from 5 Indian banks (HDFC, ICICI, PNB, SBI, Axis) plus RBI regulatory references.
2. Splits each document into semantically meaningful chunks (numbered sections or Q/A pairs).
3. Builds two parallel retrieval indexes — a dense **FAISS** vector index and a sparse **BM25** keyword index.
4. Fuses both retrieval signals with **Reciprocal Rank Fusion (RRF)**, then **re-ranks** with a cross-encoder for precision.
5. Feeds the top passages to a local, no-API-key LLM (**Qwen2.5-1.5B-Instruct**) with strict grounding instructions so it never hallucinates numbers.
6. Wraps everything in a **Gradio chat UI**, with a deterministic EMI calculator shortcut so simple math never depends on the LLM.

**Design philosophy:** this is a *closed-book* assistant — it must never answer from general knowledge, must never mix figures across banks/products, and must refuse or flag uncertainty rather than guess. This matters because giving a wrong CIBIL score or EMI figure to a real user could lead to a bad financial decision.

> **Run order matters.** Execute cells top to bottom, once, right after **Runtime → Restart session**. Several cells (chunking, embedding, indexing) create global variables (`chunks`, `model`, `index`, `bm25`, `llm_model`, `reranker`) that later cells depend on.

## Step 2: Chunking (handles numbered sections and Q/A-style FAQ docs)
Loan policy PDFs aren't uniformly structured, so this step uses a **cascading strategy** (`smart_split`) rather than one fixed parser:

1. **`split_into_sections`** — the primary strategy. Looks for numbered headings like `1. Eligibility Criteria` (regex: `^\d{1,2}\.\s+[A-Z]...`) and groups the text under each heading into one chunk. Used whenever a document has more than one such heading.
2. **`split_faq_style`** — a fallback specifically for `Loan_FAQ` documents, which are structured as `Q1. ...` / `A1. ...` pairs grouped under category headings (e.g. "Eligibility — Q3"). Keeps each question with its answer as a single chunk.
3. **`split_generic_headings`** — a last-resort fallback for documents with neither numbered sections nor Q/A pairs, grouping text under any capitalized heading-like line.

Each chunk is stored as a JSON record with rich metadata: `chunk_id`, `bank`, `doc_type` (`real_bank` vs `regulatory`), `product` (e.g. "Home Loan", "All Products", "Regulatory", "Meta"), `doc_title`, `section`, `source_file`, and `text`. This metadata is what later powers bank/product filtering in retrieval — without it, hybrid search would have no way to narrow candidates before scoring.

Chunks shorter than 15 characters are dropped as noise (page headers, stray whitespace). The final print-out shows total chunk count, a per-bank breakdown, and chunk-length statistics (min/max/mean/p90) — useful for sanity-checking that no document silently failed to parse.


```python
# Step 2: Chunking script (handles both numbered sections and Q/A-style FAQ docs)
import re, json, os

CORPUS_ROOT = "/content/corpus/data/documents"
OUT_PATH = "/content/corpus/chunks.jsonl"

SECTION_HEADING_RE = re.compile(r"^\d{1,2}\.\s+[A-Z].{2,80}$")
QA_RE = re.compile(r"^Q(\d+)\.\s+(.+)$")
A_RE = re.compile(r"^A(\d+)\.\s+(.+)$")
CATEGORY_RE = re.compile(r"^[A-Z][A-Za-z '&/()-]{2,60}$")

DOC_TITLE_MAP = {
    "Home_Loan_Policy": "Home Loan Policy", "Personal_Loan_Policy": "Personal Loan Policy",
    "Car_Loan_Policy": "Car Loan Policy", "Education_Loan_Policy": "Education Loan Policy",
    "Gold_Loan_Policy": "Gold Loan Policy", "Interest_Rate_Policy": "Interest Rate Policy",
    "Eligibility_Guide": "Eligibility Guide", "EMI_Guide": "EMI Guide",
    "Credit_Score_Guide": "Credit Score Guide", "Required_Documents": "Required Documents",
    "Loan_Application_Process": "Loan Application Process", "Loan_FAQ": "Loan FAQ",
    "Repayment_Rules": "Repayment Rules", "Foreclosure_Rules": "Foreclosure Rules",
    "Sources_and_Data_Gaps": "Sources and Data Gaps",
}
PRODUCT_FROM_DOC = {
    "Home_Loan_Policy": "Home Loan", "Personal_Loan_Policy": "Personal Loan",
    "Car_Loan_Policy": "Car Loan", "Education_Loan_Policy": "Education Loan",
    "Gold_Loan_Policy": "Gold Loan", "Interest_Rate_Policy": "All Products",
    "Eligibility_Guide": "All Products", "EMI_Guide": "All Products",
    "Credit_Score_Guide": "All Products", "Required_Documents": "All Products",
    "Loan_Application_Process": "All Products", "Loan_FAQ": "All Products",
    "Repayment_Rules": "All Products", "Foreclosure_Rules": "All Products",
    "Sources_and_Data_Gaps": "Meta",
}

def extract_pdf_text(filepath):
    import pdfplumber
    with pdfplumber.open(filepath) as pdf:
        return "\n".join(page.extract_text() or "" for page in pdf.pages)

def split_into_sections(text):
    lines = text.split("\n")
    sections, current_heading, current_body = [], "Document Header", []
    for line in lines:
        stripped = line.strip()
        if SECTION_HEADING_RE.match(stripped):
            if current_body:
                sections.append((current_heading, "\n".join(current_body).strip()))
            current_heading, current_body = stripped, []
        else:
            current_body.append(line)
    if current_body:
        sections.append((current_heading, "\n".join(current_body).strip()))
    return [(h, b) for h, b in sections if len(b) > 15]

def split_faq_style(text):
    lines = text.split("\n")
    start = 0
    for i, l in enumerate(lines):
        if l.strip().startswith("Document Code:"):
            start = i + 1
            break
    lines = lines[start:]
    chunks, current_category, current_q, current_body = [], "General", None, []
    def flush():
        if current_q and current_body:
            chunks.append((f"{current_category} — {current_q}", "\n".join(current_body).strip()))
    for line in lines:
        stripped = line.strip()
        if not stripped:
            continue
        if QA_RE.match(stripped):
            flush(); current_q, current_body = stripped, [stripped]
        elif A_RE.match(stripped):
            current_body.append(stripped)
        elif CATEGORY_RE.match(stripped) and not stripped.startswith("Disclaimer"):
            flush(); current_category, current_q, current_body = stripped, None, []
        elif current_q:
            current_body.append(stripped)
    flush()
    return [(h, b) for h, b in chunks if len(b) > 15]

def split_generic_headings(text):
    lines = text.split("\n")
    start = 0
    for i, l in enumerate(lines):
        if l.strip().startswith("Document Code:"):
            start = i + 1
            break
    lines = lines[start:]
    sections, current_heading, current_body = [], "Document Header", []
    for line in lines:
        stripped = line.strip()
        if CATEGORY_RE.match(stripped) and not stripped.startswith("Disclaimer"):
            if current_body:
                sections.append((current_heading, "\n".join(current_body).strip()))
            current_heading, current_body = stripped, []
        else:
            current_body.append(line)
    if current_body:
        sections.append((current_heading, "\n".join(current_body).strip()))
    return [(h, b) for h, b in sections if len(b) > 15]

def smart_split(text, doc_key):
    primary = split_into_sections(text)
    if len(primary) > 1:
        return primary
    if doc_key == "Loan_FAQ":
        faq = split_faq_style(text)
        if faq:
            return faq
    fallback = split_generic_headings(text)
    return fallback if fallback else primary

chunks, chunk_id = [], 0
for bank_folder in sorted(os.listdir(CORPUS_ROOT)):
    bank_dir = os.path.join(CORPUS_ROOT, bank_folder)
    if not os.path.isdir(bank_dir):
        continue
    doc_type = "regulatory" if bank_folder == "RBI" else "real_bank"
    for filename in sorted(os.listdir(bank_dir)):
        if not filename.endswith(".pdf"):
            continue
        filepath = os.path.join(bank_dir, filename)
        doc_key = filename.replace(".pdf", "")
        doc_title = DOC_TITLE_MAP.get(doc_key, doc_key.replace("_", " "))
        product = PRODUCT_FROM_DOC.get(doc_key, "Regulatory")
        text = extract_pdf_text(filepath)
        sections = smart_split(text, doc_key)
        for heading, body in sections:
            chunk_id += 1
            chunks.append({
                "chunk_id": f"chunk_{chunk_id:04d}", "bank": bank_folder, "doc_type": doc_type,
                "product": product, "doc_title": doc_title, "section": heading,
                "source_file": f"{bank_folder}/{filename}", "text": body,
            })

with open(OUT_PATH, "w") as f:
    for c in chunks:
        f.write(json.dumps(c, ensure_ascii=False) + "\n")

print(f"Total chunks: {len(chunks)}")
from collections import Counter
print(Counter(c["bank"] for c in chunks))

lens = [len(c["text"]) for c in chunks]
import numpy as np
print(f"\nchunk length: min={min(lens)} max={max(lens)} mean={np.mean(lens):.0f} p90={np.percentile(lens,90):.0f}")
```

## Step 3: Generate embeddings and build a FAISS index
Uses **`BAAI/bge-small-en-v1.5`** — chosen over MiniLM because BGE models are explicitly tuned for retrieval tasks and generally outperform MiniLM-based embedders on passage-ranking benchmarks, at a similar size/speed cost.

Each chunk is embedded with its metadata prefixed onto the text (`"{bank} | {doc_title} | {section}\n{text}"`), not just the raw passage. This gives the embedding model explicit signal about which bank/document/section a chunk belongs to, which helps dense retrieval distinguish otherwise similar-sounding passages from different banks.

- `normalize_embeddings=True` — vectors are L2-normalized, so...
- `faiss.IndexFlatIP` — ...inner product search is equivalent to cosine similarity. This is an *exact* (brute-force) index, appropriate at this corpus scale (a few thousand chunks); it trades a small amount of speed for perfect recall versus an approximate index like HNSW/IVF.

Artifacts persisted to disk for reuse: `faiss_index.bin` (the index itself), `embeddings.npy` (raw vectors), and `chunk_ids.json` (the ID ordering that maps index positions back to chunks).


```python
# Step 3: Generate embeddings and build a FAISS index
from sentence_transformers import SentenceTransformer
import faiss, numpy as np, json

model = SentenceTransformer("BAAI/bge-small-en-v1.5")   # stronger than MiniLM for retrieval

texts_to_embed = [f"{c['bank']} | {c['doc_title']} | {c['section']}\n{c['text']}" for c in chunks]

embeddings = model.encode(texts_to_embed, show_progress_bar=True, convert_to_numpy=True, normalize_embeddings=True)

dimension = embeddings.shape[1]
index = faiss.IndexFlatIP(dimension)
index.add(embeddings)

faiss.write_index(index, "/content/corpus/faiss_index.bin")
np.save("/content/corpus/embeddings.npy", embeddings)

chunk_ids = [c["chunk_id"] for c in chunks]
with open("/content/corpus/chunk_ids.json", "w") as f:
    json.dump(chunk_ids, f)

print(f"Indexed {index.ntotal} chunks, dim={dimension}")
```

## Step 4: Build BM25 index
BM25 is a classic sparse lexical scorer — it complements the dense FAISS index by catching **exact keyword matches** (bank names, precise terms like "CIBIL", "foreclosure") that embeddings can sometimes under-weight in favor of semantic similarity.

Tokenization here is intentionally simple — lowercase + whitespace split (`c["text"].lower().split()`) — no stemming or stopword removal, which keeps behavior predictable and matches how `bm25.get_scores()` is called elsewhere in the notebook (same tokenization must be used at query time for scores to be meaningful).

The index is pickled to `bm25.pkl` for reuse, and a quick sanity check runs the query *"minimum CIBIL score"* through it, printing the top 5 scoring chunks — a fast way to confirm the index is actually surfacing relevant text before building anything on top of it.

```python
# Step 4: Build BM25 index
from rank_bm25 import BM25Okapi
import pickle

tokenized_corpus = [c["text"].lower().split() for c in chunks]
bm25 = BM25Okapi(tokenized_corpus)

with open("/content/corpus/bm25.pkl", "wb") as f:
    pickle.dump(bm25, f)

print(f"BM25 index built over {len(tokenized_corpus)} chunks")

# quick sanity check — a keyword-heavy query should surface relevant chunks
test_query = "minimum CIBIL score"
scores = bm25.get_scores(test_query.lower().split())
top5 = sorted(range(len(scores)), key=lambda i: -scores[i])[:5]
for i in top5:
    print(f"{scores[i]:.2f} | {chunks[i]['bank']} | {chunks[i]['doc_title']} | {chunks[i]['section']}")
```

## Step 5: Metadata filter extraction (bank / product keyword detection)
Before scoring anything, it's cheaper and more precise to first **narrow the candidate pool** using simple keyword matching against the user's query:

- **`BANKS`** — a fixed allow-list (`HDFC, ICICI, PNB, SBI, Axis`). If any of these appears in the query, retrieval later restricts to that bank's chunks (plus RBI, which is always eligible since it's regulatory, not bank-specific).
- **`PRODUCT_KEYWORDS`** — maps loose everyday phrasing ("housing", "auto", "vehicle", "student") to the canonical product labels used in chunk metadata ("Home Loan", "Car Loan", "Education Loan", etc.).

`extract_filters()` returns a `(bank, product)` tuple, with `None` for anything not detected — the sanity-check calls at the bottom show this correctly picks up an explicit bank+product mention, a product-only mention, and a query with neither (an RBI/regulatory question).


```python
# Step 5: Metadata filter extraction
BANKS = ["HDFC", "ICICI", "PNB", "SBI", "Axis"]
PRODUCT_KEYWORDS = {
    "home": "Home Loan", "housing": "Home Loan",
    "personal": "Personal Loan",
    "car": "Car Loan", "auto": "Car Loan", "vehicle": "Car Loan",
    "education": "Education Loan", "student": "Education Loan",
    "gold": "Gold Loan",
}

def extract_filters(query):
    q = query.lower()
    bank = next((b for b in BANKS if b.lower() in q), None)
    product = next((v for k, v in PRODUCT_KEYWORDS.items() if k in q), None)
    return bank, product

# sanity check
print(extract_filters("What is the interest rate for HDFC home loan?"))
print(extract_filters("gold loan eligibility at SBI"))
print(extract_filters("what does RBI say about penal charges?"))
```

## Step 5b : Hybrid retrieval — filter, then dense + BM25 fused by reciprocal rank
`hybrid_search()` is the core retrieval function, combining three ideas:

1. **Metadata pre-filtering** — restricts candidates to the detected bank (plus RBI) and product (plus "All Products"/"Regulatory"/"Meta") from Step 5. If filtering leaves zero candidates (e.g. filters were too aggressive or nothing detected), it safely falls back to the full corpus rather than returning nothing.
2. **Dense + sparse scoring in parallel** — the query is embedded and searched against the FAISS index for a full ranking (`dense_rank`), while BM25 scores every candidate chunk independently (`bm25_rank`).
3. **Reciprocal Rank Fusion (RRF)** — rather than trying to normalize and average two differently-scaled scores, RRF combines *ranks*: `1/(k_rrf + rank)` for each method, summed per chunk. This is a well-established, parameter-light way to fuse heterogeneous retrievers (`k_rrf=60` is a standard damping constant from the original RRF paper) without needing to tune score weightings.

Chunks not found in one ranking (e.g. filtered out) get a large default rank (`10_000`), effectively giving them a near-zero contribution from that method rather than crashing on a missing key.

The test call at the bottom retrieves for *"what is the minimum CIBIL score for HDFC home loan"* and prints bank/document/section/text-snippet for each hit — a fast visual check that filtering + fusion is surfacing HDFC-specific, home-loan-relevant chunks rather than generic or cross-bank noise.


```python
# Step 5b: Hybrid retrieval — filter, then dense + BM25 fused by reciprocal rank
def hybrid_search(query, top_k=5, k_rrf=60):
    bank, product = extract_filters(query)

    candidate_idx = [
        i for i, c in enumerate(chunks)
        if (bank is None or c["bank"] == bank or c["bank"] == "RBI")
        and (product is None or c["product"] in (product, "All Products", "Regulatory", "Meta"))
    ]
    if not candidate_idx:
        candidate_idx = list(range(len(chunks)))

    q_emb = model.encode([query], convert_to_numpy=True, normalize_embeddings=True)
    dense_scores, dense_idx = index.search(q_emb, len(chunks))
    dense_rank = {idx: r for r, idx in enumerate(dense_idx[0]) if idx in candidate_idx}

    bm25_scores = bm25.get_scores(query.lower().split())
    bm25_order = sorted(candidate_idx, key=lambda i: -bm25_scores[i])
    bm25_rank = {idx: r for r, idx in enumerate(bm25_order)}

    fused = {}
    for idx in candidate_idx:
        fused[idx] = 1/(k_rrf + dense_rank.get(idx, 10_000)) + 1/(k_rrf + bm25_rank.get(idx, 10_000))

    top = sorted(fused, key=fused.get, reverse=True)[:top_k]
    return [chunks[i] for i in top]

# test it
for r in hybrid_search("what is the minimum CIBIL score for HDFC home loan"):
    print(r["bank"], "|", r["doc_title"], "|", r["section"], "->", r["text"][:100].replace(chr(10)," "))
```

## Step 7: Cross-encoder re-ranker
`hybrid_search` (Step 5b) is fast but approximate — it fuses two independent rankings without ever letting the query and passage "see" each other directly. A **cross-encoder** fixes that by scoring `(query, passage)` pairs jointly, which is far more accurate at judging true relevance, at the cost of being too slow to run over the whole corpus.

So the pattern here is the standard **retrieve-then-re-rank** pipeline:

1. `hybrid_search` pulls a wider candidate set (`retrieve_k=20`, i.e. cast a wide net cheaply).
2. **`cross-encoder/ms-marco-MiniLM-L-6-v2`** scores each of those 20 candidates against the query.
3. Candidates are re-sorted by cross-encoder score, and only the top `final_k=5` are kept.

`retrieve_with_rerank()` wraps this whole flow and attaches the resulting `rerank_score` onto each returned chunk — this score is reused later (Step 8) as the confidence signal that decides whether to answer or refuse.


```python
!pip install -q sentence-transformers

from sentence_transformers import CrossEncoder

reranker = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")
print("Cross-encoder re-ranker loaded.")

def retrieve_with_rerank(query, retrieve_k=20, final_k=5):
    candidates = hybrid_search(query, top_k=retrieve_k)
    candidate_chunks = [chunks[r] if isinstance(r, int) else r for r in candidates]

    pairs = [(query, c["text"]) for c in candidate_chunks]
    scores = reranker.predict(pairs)

    reranked = sorted(zip(candidate_chunks, scores), key=lambda x: x[1], reverse=True)
    top = reranked[:final_k]

    results = []
    for c, score in top:
        c_copy = dict(c)
        c_copy["rerank_score"] = float(score)
        results.append(c_copy)
    return results
```

## Step 8: Answer generation — single, final version
This cell intentionally consolidates *every* piece of answer-generation logic — loan-type filtering, uncertainty flagging, unknown-bank detection, confidence-threshold refusal — into **one definitive set of function definitions**, so re-running earlier draft cells out of order can never silently shadow part of the logic. Key pieces:

- **`SYSTEM_PROMPT`** — an unusually strict grounding prompt: answer only from provided sources, cite with `[1]`, `[2]`, never invent numbers, never average conflicting bank figures, never mix loan products, never give a yes/no eligibility verdict unless a source states that exact threshold. This is the main defense against hallucinated financial figures.
- **`CONFIDENCE_THRESHOLD = 1.0`** — if the top re-rank score falls below this, the pipeline refuses outright rather than attempting an answer with weak evidence.
- **`check_unknown_entity()`** — regex-detects a `"<Name> Bank"` / `"Bank <Name>"` pattern in the query and checks it against `KNOWN_BANKS`; catches questions about banks that simply aren't in the corpus (e.g. "Bank XYZ") so the system says so explicitly instead of quietly answering from the wrong bank's documents.
- **`get_loan_type()` / `filter_relevant_chunks()`** — detects which loan product the question is about and keeps only chunks mentioning ≥2 relevant keywords, so a home-loan question can't accidentally surface personal-loan figures.
- **`has_uncertainty_flag()`** — scans retrieved text for markers like *"not confidently found"* or *"data gap"* (phrases the source corpus itself uses to flag unreliable data) and, if present, injects extra instructions telling the LLM to surface the conflict rather than pick one number.
- **`is_comparison_query()` / `bank_filtered_retrieve()` / `retrieve_for_comparison()`** — for "compare X vs Y" questions, retrieval runs **separately per named bank** (bypassing the shared filter) and results are merged. This prevents one bank's chunks from crowding out another's purely because they scored higher in a single fused ranking.
- **`generate_answer_local()`** — assembles the numbered source block, builds a detailed "STRICT ANSWERING RULES" user message, applies the tokenizer's chat template, and generates with `do_sample=False` (fully deterministic/greedy decoding — appropriate here since consistency matters more than creative variation for financial answers).
- **`answer_query()`** — the public entry point. Order of checks: unknown-bank refusal → comparison-query path (skips the single confidence check, since comparisons are validated per-bank) → confidence-threshold refusal → normal generation.

```python
import re

SYSTEM_PROMPT = (
    "You are a loan advisory assistant for Indian banks. Answer ONLY using the "
    "provided source excerpts below — never from general knowledge. "
    "Cite each claim inline like [1], [2] matching the source numbers. "
    "If the sources don't contain enough information to answer, say so explicitly "
    "instead of guessing. If sources say a value is 'not confidently found', say "
    "that value is not officially confirmed rather than inventing a number. "
    "If different banks give different figures, list them separately rather than "
    "averaging or picking one. "
    "Never use your own knowledge. If a fact is not explicitly written in the "
    "retrieved sources, say the information is unavailable — do not infer it. "
    "Do not combine information from different loan products. Do not combine "
    "information from different banks unless the user explicitly asked for a "
    "comparison. Never give a yes/no verdict on whether someone qualifies for a "
    "loan unless a source states that exact threshold explicitly — if the "
    "sources only give a range or a different bank's figure, present the range "
    "and say eligibility for that specific score is not stated, rather than "
    "answering yes or no. Do not attribute a number to a bank unless that "
    "number appears in a source explicitly labeled with that bank. Answer "
    "naturally, like a chatbot — avoid opening with 'According to the sources' "
    "or 'According to the document' unless it's necessary for clarity."

)

CONFIDENCE_THRESHOLD = 1.0  # below this, refuse rather than guess

KNOWN_BANKS = set(BANKS) | {"RBI"}

UNCERTAINTY_MARKERS = [
    "not confidently found",
    "conflicted",
    "verify directly",
    "data gap",
    "not officially confirmed",
    "not confidently confirmed",
    "sources conflicted",
    "unclear",
]

COMPARISON_MARKERS = ["compare", " vs ", " vs.", "versus", "compare to", "difference between"]


def format_sources(retrieved_chunks):
    blocks = []
    for i, c in enumerate(retrieved_chunks, 1):
        blocks.append(
            f"[{i}] Bank: {c['bank']} | Document: {c['doc_title']} | Section: {c['section']}\n{c['text']}"
        )
    return "\n\n".join(blocks)


def has_uncertainty_flag(chunks_list):
    if not chunks_list:
        return False
    combined = " ".join(c["text"].lower() for c in chunks_list)
    return any(marker in combined for marker in UNCERTAINTY_MARKERS)


def check_unknown_entity(query):
    """Catch questions that name a specific bank not present in the corpus."""
    matches = re.findall(r"\b([A-Z][a-zA-Z]+)\s+Bank\b|\bBank\s+([A-Z][a-zA-Z]+)\b", query)
    named = set()
    for a, b in matches:
        if a: named.add(a)
        if b: named.add(b)
    unknown = [n for n in named if n not in KNOWN_BANKS]
    return unknown


def get_loan_type(query):
    query = query.lower()
    if "home loan" in query:
        return "home loan"
    elif "personal loan" in query:
        return "personal loan"
    elif "car loan" in query:
        return "car loan"
    elif "education loan" in query:
        return "education loan"
    elif "gold loan" in query:
        return "gold loan"
    return None


def filter_relevant_chunks(query, chunks_in):
    loan_type = get_loan_type(query)
    if loan_type is None:
        return chunks_in

    keyword_map = {
        "home loan": ["home loan", "home loans", "housing", "housing loan", "floating",
                      "fixed rate", "individual borrower", "individual-purpose", "foreclosure"],
        "personal loan": ["personal loan", "foreclosure"],
        "car loan": ["car loan", "vehicle loan", "foreclosure"],
        "education loan": ["education loan", "foreclosure"],
        "gold loan": ["gold loan", "foreclosure"],
    }
    keywords = keyword_map[loan_type]

    scored_chunks = []
    for chunk in chunks_in:
        text = chunk["text"].lower()
        score = sum(1 for word in keywords if word in text)
        if score >= 2:
            scored_chunks.append(chunk)

    return scored_chunks if scored_chunks else chunks_in


def is_comparison_query(query):
    q = query.lower()
    return any(marker in q for marker in COMPARISON_MARKERS)


def banks_named_in_query(query):
    q = query.lower()
    return [b for b in BANKS if b.lower() in q]


def bank_filtered_retrieve(query, bank, top_k=4, retrieve_k=20, k_rrf=60):
    """
    Retrieve top_k chunks belonging ONLY to the given bank (or RBI regulatory docs).
    Bypasses extract_filters() entirely, so it can't be hijacked by another bank's
    name appearing earlier in the query text.
    """
    candidate_idx = [
        i for i, c in enumerate(chunks)
        if c["bank"] == bank or c["bank"] == "RBI"
    ]
    if not candidate_idx:
        return []

    q_emb = model.encode([query], convert_to_numpy=True, normalize_embeddings=True)
    dense_scores, dense_idx = index.search(q_emb, len(chunks))
    dense_rank = {idx: r for r, idx in enumerate(dense_idx[0]) if idx in candidate_idx}

    bm25_scores = bm25.get_scores(query.lower().split())
    bm25_order = sorted(candidate_idx, key=lambda i: -bm25_scores[i])
    bm25_rank = {idx: r for r, idx in enumerate(bm25_order)}

    fused = {}
    for idx in candidate_idx:
        fused[idx] = 1 / (k_rrf + dense_rank.get(idx, 10_000)) + 1 / (k_rrf + bm25_rank.get(idx, 10_000))

    top = sorted(fused, key=fused.get, reverse=True)[:retrieve_k]
    candidates = [chunks[i] for i in top]

    pairs = [(query, c["text"]) for c in candidates]
    scores = reranker.predict(pairs)
    reranked = sorted(zip(candidates, scores), key=lambda x: x[1], reverse=True)[:top_k]

    results = []
    for c, score in reranked:
        c_copy = dict(c)
        c_copy["rerank_score"] = float(score)
        results.append(c_copy)
    return results


def retrieve_for_comparison(query, per_bank_k=4):
    """
    For 'compare X vs Y' style queries, retrieve separately per named bank using
    bank_filtered_retrieve, so one bank's chunks can never crowd out another's.
    If fewer than 2 banks are named, samples across all banks.
    """
    named = banks_named_in_query(query)
    if len(named) < 2:
        named = BANKS  # no specific banks named — sample every bank

    all_chunks = []
    seen_ids = set()
    for bank in named:
        results = bank_filtered_retrieve(query, bank, top_k=per_bank_k)
        for c in results:
            if c["chunk_id"] not in seen_ids:
                all_chunks.append(c)
                seen_ids.add(c["chunk_id"])

    all_chunks.sort(key=lambda c: c.get("rerank_score", 0), reverse=True)
    return all_chunks


def generate_answer_local(query, top_k=5, max_new_tokens=220):
    if is_comparison_query(query):
        retrieved_chunks = retrieve_for_comparison(query)
    else:
        retrieved_chunks = retrieve_with_rerank(query, retrieve_k=20, final_k=top_k)
        retrieved_chunks = filter_relevant_chunks(query, retrieved_chunks)

    # Improvement 1: guard against empty retrieval
    if not retrieved_chunks:
        return {
            "query": query,
            "answer": "I couldn't find any relevant information in the available loan documents.",
            "citations": [],
            "top_score": None,
            "refused": True,
            "flagged_uncertain": False,
        }

    sources_block = format_sources(retrieved_chunks)
    uncertain = has_uncertainty_flag(retrieved_chunks)

    uncertainty_instruction = ""
    if uncertain:
        uncertainty_instruction = """

IMPORTANT:
The retrieved documents contain conflicting, incomplete, or unverified information.

Follow these rules:
- Clearly mention that the sources conflict.
- Do not choose one value as the final truth.
- Keep every number attached to its condition.
- Do not merge different loan categories.
- Recommend verification with the official bank source.
"""

    user_message = f"""
Question:
{query}

Sources:
{sources_block}

STRICT ANSWERING RULES:
1. Answer ONLY the question asked.
2. Discuss ONLY the loan category mentioned in the question, unless the question
   is explicitly a comparison across banks. Never mention unrelated products
   (e.g. if asked about Home Loan, do not discuss Personal/Car/Gold/Education Loan).
3. Use ONLY information present in the sources. Do NOT use general banking knowledge.
4. Do NOT guess missing values or average percentages.
5. Never mix conditions — keep each figure attached to the condition it belongs to.
   Example: "Business-purpose loans mention 3-4%, while floating-rate individual loans mention NIL."
   NOT: "Floating loans have 3-4% charges."
6. If information is unavailable, say that it is not available in the documents.
7. Never attribute a number to a bank unless that exact source is labeled with that bank.
8. If asked whether a specific score/value qualifies (e.g. "can I get a loan with score 500"),
   do NOT answer yes or no unless a source states that exact value's outcome. Otherwise state
   the range given in the sources and say that specific value's outcome isn't stated.
9. Add citation numbers like [1], [2].
{uncertainty_instruction}
"""

    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": user_message},
    ]

    prompt_text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(prompt_text, return_tensors="pt").to(llm_model.device)

    with torch.no_grad():
        output_ids = llm_model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id,
        )

    answer_text = tokenizer.decode(
        output_ids[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True
    ).strip()

    citations = [
        {"ref": i + 1, "bank": c["bank"], "doc": c["doc_title"], "section": c["section"],
         "rerank_score": c.get("rerank_score")}
        for i, c in enumerate(retrieved_chunks)
    ]

    return {
        "query": query,
        "answer": answer_text,
        "citations": citations,
        "top_score": retrieved_chunks[0].get("rerank_score", 0),
        "flagged_uncertain": uncertain,
    }


def answer_query(query, top_k=5, max_new_tokens=220):
    unknown_entities = check_unknown_entity(query)
    if unknown_entities:
        return {
            "query": query,
            "answer": f"I don't have documents covering {', '.join(unknown_entities)}. "
                      f"I can answer questions about: {', '.join(sorted(KNOWN_BANKS))}.",
            "citations": [],
            "top_score": None,
            "refused": True,
            "flagged_uncertain": False,
        }

    # comparison queries skip the single-topline confidence check — they're
    # evaluated per-bank inside generate_answer_local instead
    if is_comparison_query(query):
        result = generate_answer_local(query, top_k=top_k, max_new_tokens=max_new_tokens)
        result["refused"] = False
        return result

    retrieved_chunks = retrieve_with_rerank(query, retrieve_k=20, final_k=top_k)

    if not retrieved_chunks:
        return {
            "query": query,
            "answer": "I couldn't find any relevant information in the available loan documents.",
            "citations": [],
            "top_score": None,
            "refused": True,
            "flagged_uncertain": False,
        }

    top_score = retrieved_chunks[0].get("rerank_score", -999)

    if top_score < CONFIDENCE_THRESHOLD:
        return {
            "query": query,
            "answer": "I couldn't find reliable information about this in the loan policy documents. "
                      "Please rephrase your question or consult your bank directly.",
            "citations": [],
            "top_score": top_score,
            "refused": True,
            "flagged_uncertain": False,
        }

    result = generate_answer_local(query, top_k=top_k, max_new_tokens=max_new_tokens)
    result["refused"] = False
    return result

print("Pipeline functions loaded: answer_query() is ready.")
```

## Step 9b: Let users upload their own documents (PDF / TXT / DOCX)
This adds any file a user uploads straight into the same FAISS + BM25 index used for the bank corpus, tagged with `bank = "User Upload"`, so `hybrid_search` / `answer_query` can retrieve from it exactly like any other document — no separate pipeline needed.

```python
# Step 9b: Ingest user-uploaded documents into the existing index
import os, re, json
import numpy as np

UPLOAD_DIR = "/content/user_uploads"
os.makedirs(UPLOAD_DIR, exist_ok=True)

# keeps track of what each user has uploaded (filename -> list of chunk_ids)
uploaded_doc_registry = {}


def extract_text_from_file(filepath):
    """Extract raw text from a pdf / txt / docx file."""
    ext = filepath.lower().rsplit(".", 1)[-1]
    if ext == "pdf":
        return extract_pdf_text(filepath)  # reuses the pdfplumber helper from Step 2
    elif ext == "txt":
        with open(filepath, "r", encoding="utf-8", errors="ignore") as f:
            return f.read()
    elif ext == "docx":
        try:
            import docx
        except ImportError:
            import subprocess
            subprocess.run(["pip", "install", "-q", "python-docx"], check=True)
            import docx
        d = docx.Document(filepath)
        return "\n".join(p.text for p in d.paragraphs)
    else:
        raise ValueError(f"Unsupported file type: .{ext} (use pdf, txt, or docx)")


def chunk_generic_text(text, max_chars=1200):
    """Simple paragraph-aware chunker for arbitrary user documents
    (the bank corpus has its own structured chunker in Step 2 — user
    uploads don't follow that structure, so we keep this generic)."""
    paragraphs = [p.strip() for p in re.split(r"\n\s*\n", text) if p.strip()]
    pieces, buf = [], ""
    for p in paragraphs:
        if len(buf) + len(p) < max_chars:
            buf = (buf + "\n\n" + p).strip()
        else:
            if buf:
                pieces.append(buf)
            buf = p
    if buf:
        pieces.append(buf)
    return pieces if pieces else ([text[:max_chars]] if text.strip() else [])


def add_uploaded_documents(file_paths):
    """
    Extract, chunk, embed, and add one or more user-uploaded files into the
    live FAISS index + BM25 index + chunks list, so they become immediately
    queryable through hybrid_search / answer_query.
    Returns the list of filenames that were successfully added.
    """
    global chunks, index, bm25, tokenized_corpus, chunk_ids

    new_chunks = []
    added_names = []

    for filepath in file_paths:
        filename = os.path.basename(filepath)
        try:
            text = extract_text_from_file(filepath)
        except Exception as e:
            print(f"Skipping {filename}: {e}")
            continue
        if not text or not text.strip():
            print(f"Skipping {filename}: no extractable text")
            continue

        pieces = chunk_generic_text(text)
        doc_chunk_ids = []
        for j, body in enumerate(pieces, 1):
            cid = f"user_{os.path.splitext(filename)[0]}_{j:03d}"
            c = {
                "chunk_id": cid,
                "bank": "User Upload",
                "doc_type": "user_upload",
                "product": "User Document",
                "doc_title": filename,
                "section": f"Part {j}",
                "source_file": filename,
                "text": body,
            }
            new_chunks.append(c)
            doc_chunk_ids.append(cid)

        uploaded_doc_registry[filename] = doc_chunk_ids
        added_names.append(filename)

    if not new_chunks:
        return []

    # embed the new chunks the same way the original corpus was embedded (Step 3)
    texts_to_embed = [
        f"{c['bank']} | {c['doc_title']} | {c['section']}\n{c['text']}" for c in new_chunks
    ]
    new_embeddings = model.encode(
        texts_to_embed, convert_to_numpy=True, normalize_embeddings=True
    )
    index.add(new_embeddings)

    chunks.extend(new_chunks)
    chunk_ids.extend([c["chunk_id"] for c in new_chunks])

    # BM25 has no incremental API — rebuild it over the (now larger) corpus.
    # This is cheap even for thousands of chunks, so it's fine to do per-upload.
    tokenized_corpus = [c["text"].lower().split() for c in chunks]
    bm25 = BM25Okapi(tokenized_corpus)

    print(f"Added {len(new_chunks)} chunks from {len(added_names)} file(s): {added_names}")
    return added_names


def remove_uploaded_document(filename):
    """Optional: remove a previously uploaded doc's chunks from the in-memory
    corpus. Note: FAISS IndexFlatIP doesn't support deleting by id cheaply, so
    this rebuilds the FAISS index and BM25 index from the filtered chunk list."""
    global chunks, index, bm25, tokenized_corpus, chunk_ids

    if filename not in uploaded_doc_registry:
        print(f"No record of {filename}")
        return

    ids_to_remove = set(uploaded_doc_registry[filename])
    chunks = [c for c in chunks if c["chunk_id"] not in ids_to_remove]
    chunk_ids = [c["chunk_id"] for c in chunks]
    del uploaded_doc_registry[filename]

    texts_to_embed = [f"{c['bank']} | {c['doc_title']} | {c['section']}\n{c['text']}" for c in chunks]
    embeddings_rebuilt = model.encode(texts_to_embed, convert_to_numpy=True, normalize_embeddings=True)
    index = faiss.IndexFlatIP(embeddings_rebuilt.shape[1])
    index.add(embeddings_rebuilt)

    tokenized_corpus = [c["text"].lower().split() for c in chunks]
    bm25 = BM25Okapi(tokenized_corpus)

    print(f"Removed {filename}. Corpus now has {len(chunks)} chunks.")

print("Upload ingestion ready: add_uploaded_documents(), remove_uploaded_document().")
```

### Step 9c: Patch `hybrid_search` so user-uploaded documents are always considered
The original `hybrid_search` (Step 5b) only includes a chunk as a candidate when its `bank` matches a bank named in the query, or is `"RBI"`. This redefinition adds `"User Upload"` to that always-included set, so uploaded documents are retrievable even when the user's question doesn't name a bank.

```python
# Step 9c: Redefine hybrid_search to always include user-uploaded chunks as candidates
def hybrid_search(query, top_k=5, k_rrf=60):
    bank, product = extract_filters(query)

    candidate_idx = [
        i for i, c in enumerate(chunks)
        if (bank is None or c["bank"] == bank or c["bank"] in ("RBI", "User Upload"))
        and (product is None or c["product"] in (product, "All Products", "Regulatory", "Meta", "User Document"))
    ]
    if not candidate_idx:
        candidate_idx = list(range(len(chunks)))

    q_emb = model.encode([query], convert_to_numpy=True, normalize_embeddings=True)
    dense_scores, dense_idx = index.search(q_emb, len(chunks))
    dense_rank = {idx: r for r, idx in enumerate(dense_idx[0]) if idx in candidate_idx}

    bm25_scores = bm25.get_scores(query.lower().split())
    bm25_order = sorted(candidate_idx, key=lambda i: -bm25_scores[i])
    bm25_rank = {idx: r for r, idx in enumerate(bm25_order)}

    fused = {}
    for idx in candidate_idx:
        fused[idx] = 1/(k_rrf + dense_rank.get(idx, 10_000)) + 1/(k_rrf + bm25_rank.get(idx, 10_000))

    top = sorted(fused, key=fused.get, reverse=True)[:top_k]
    return [chunks[i] for i in top]

print("hybrid_search patched to always include User Upload chunks.")
```

## Step 10: Gradio chat UI
### With a deterministic, rule-based EMI calculator shortcut

Wraps the pipeline in an interactive `gr.ChatInterface`. The notable design choice here: **EMI math is never delegated to the LLM.**

- **`calculate_emi()`** — the standard amortizing-loan EMI formula: `EMI = P × r × (1+r)^n / ((1+r)^n − 1)`, with a zero-interest fallback (`P/n`) to avoid a division-by-zero. Also returns total interest and total payment.
- **`try_parse_emi_request()`** — before touching the LLM, detects EMI-calculation intent from keywords ("emi", "calculate", "monthly installment"/"payment") and extracts three numbers from free text using regex: loan amount (handling Indian numeral shorthand like *"10 lakh"*, *"1 crore"*, or plain digits/commas), interest rate (`%`/"percent"), and tenure (years or months). Returns `None` if any of the three is missing, so the caller can fall through to asking the user for the missing piece rather than guessing.
- **`chat_fn()`** — the routing logic: (1) if all three EMI inputs are present, compute and return the exact figure directly — bypassing the LLM entirely, so EMI numbers are always arithmetically exact rather than model-generated; (2) if EMI intent is detected but a number is missing, ask the user to supply it; (3) otherwise, route to `answer_query()` (the RAG pipeline) and append a formatted **Sources** list built from the returned citations.

`demo.launch(share=True, debug=True)` produces a public shareable Gradio link and keeps debug logging on — useful while testing, but worth turning `share=False`/`debug=False` off before any real deployment.

```python
!pip install -q gradio
import re
import os
import gradio as gr

# ---------- EMI calculation logic ----------
def calculate_emi(principal, annual_rate, tenure_years):
    monthly_rate = (annual_rate / 12) / 100
    tenure_months = int(tenure_years * 12)

    if monthly_rate == 0:
        emi = principal / tenure_months
    else:
        emi = principal * monthly_rate * (1 + monthly_rate) ** tenure_months / \
              ((1 + monthly_rate) ** tenure_months - 1)

    total_payment = emi * tenure_months
    total_interest = total_payment - principal
    return emi, total_interest, total_payment


# ---------- Detect an EMI calculation request and pull out numbers ----------
def try_parse_emi_request(message):
    msg = message.lower()
    if not any(k in msg for k in ["emi", "calculate", "monthly installment", "monthly payment"]):
        return None

    amount = None
    lakh_match = re.search(r"(\d+(?:\.\d+)?)\s*lakh", msg)
    crore_match = re.search(r"(\d+(?:\.\d+)?)\s*crore", msg)
    plain_amount_match = re.search(r"(?:rs\.?|₹|inr)?\s*([\d,]{5,})", msg)

    if crore_match:
        amount = float(crore_match.group(1)) * 1_00_00_000
    elif lakh_match:
        amount = float(lakh_match.group(1)) * 1_00_000
    elif plain_amount_match:
        amount = float(plain_amount_match.group(1).replace(",", ""))

    rate_match = re.search(r"(\d+(?:\.\d+)?)\s*(?:%|percent)", msg)
    rate = float(rate_match.group(1)) if rate_match else None

    years_match = re.search(r"(\d+(?:\.\d+)?)\s*year", msg)
    months_match = re.search(r"(\d+(?:\.\d+)?)\s*month", msg)
    tenure_years = None
    if years_match:
        tenure_years = float(years_match.group(1))
    elif months_match:
        tenure_years = float(months_match.group(1)) / 12

    if amount and rate and tenure_years:
        return amount, rate, tenure_years
    return None


SUPPORTED_UPLOAD_EXTS = (".pdf", ".txt", ".docx")

# ---------- Combined chat handler (now multimodal: text + file attachments) ----------
def chat_fn(message, history):
    # gr.ChatInterface(multimodal=True) passes message as
    # {"text": str, "files": [filepath, ...]}
    if isinstance(message, dict):
        text = (message.get("text") or "").strip()
        files = message.get("files") or []
    else:
        text = (message or "").strip()
        files = []

    upload_note = ""
    if files:
        valid_files = [f for f in files if f.lower().endswith(SUPPORTED_UPLOAD_EXTS)]
        skipped = [os.path.basename(f) for f in files if f not in valid_files]

        added = add_uploaded_documents(valid_files) if valid_files else []

        parts = []
        if added:
            parts.append(f"📄 Added to the knowledge base: {', '.join(added)}. You can now ask questions about it.")
        if skipped:
            parts.append(f"⚠️ Skipped unsupported file(s): {', '.join(skipped)} (use PDF, TXT, or DOCX).")
        upload_note = "\n\n".join(parts)

        if not text:
            # user only uploaded files, no question yet
            return upload_note or "I couldn't read that file — please try a PDF, TXT, or DOCX."

    if not text:
        return "Please type a question, or attach a PDF/TXT/DOCX and ask something about it."

    emi_params = try_parse_emi_request(text)
    if emi_params:
        principal, rate, tenure = emi_params
        emi, total_interest, total_payment = calculate_emi(principal, rate, tenure)
        answer = (
            f"**EMI Calculation**\n\n"
            f"- Loan Amount: ₹{principal:,.0f}\n"
            f"- Interest Rate: {rate}% per annum\n"
            f"- Tenure: {tenure:.0f} years\n\n"
            f"**Monthly EMI: ₹{emi:,.2f}**\n"
            f"Total Interest Payable: ₹{total_interest:,.2f}\n"
            f"Total Payment: ₹{total_payment:,.2f}"
        )
        return (upload_note + "\n\n" + answer) if upload_note else answer

    msg_lower = text.lower()
    if any(k in msg_lower for k in ["emi", "calculate emi", "monthly installment"]) and re.search(r"\d", text):
        answer = ("To calculate your EMI, please give me all three: loan amount, interest rate, and tenure.\n"
                   "Example: *\"Calculate EMI for 10 lakh at 8.5% for 20 years\"*")
        return (upload_note + "\n\n" + answer) if upload_note else answer

    # otherwise, route to the RAG pipeline (now also searches any uploaded docs)
    result = answer_query(text)
    answer = result["answer"]
    if not result["refused"] and result["citations"]:
        sources_text = "\n\n**Sources:**\n" + "\n".join(
            f"[{c['ref']}] {c['bank']} — {c['doc']} — {c['section']}"
            for c in result["citations"]
        )
        answer += sources_text

    return (upload_note + "\n\n" + answer) if upload_note else answer


demo = gr.ChatInterface(
    fn=chat_fn,
    multimodal=True,  # adds a paperclip button to the chat box for file uploads
    title="Loan Advisory Assistant",
    description=(
        "Ask about loan eligibility, EMIs, documentation, foreclosure charges, RBI rules — or ask me to "
        "calculate your EMI directly (e.g. \"Calculate EMI for 10 lakh at 8.5% for 20 years\").\n\n"
        "📎 You can also attach your own PDF, TXT, or DOCX file (use the paperclip icon) — I'll add it to "
        "the knowledge base and you can then ask questions about it."
    ),
    examples=[
        {"text": "What is the minimum CIBIL score for a home loan?"},
        {"text": "Calculate EMI for 10 lakh at 8.5% for 20 years"},
        {"text": "What documents do I need for a personal loan?"},
        {"text": "What is RBI's penal charge policy?"},
    ],
)

demo.launch(share=True, debug=True)
```
